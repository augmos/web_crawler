```python
keyword = "Matrix"

website = "youtube_channel" # youtube, youtube_channel, medium, google
channel="beckuplearning" #falls "youtube_channel" ausgewählt
                        #beckuplearning= Daniel Jung, TheSimpleMaths = The Simple Club

if website == "youtube":
    link="https://youtube.com/results?search_query=" # youtube search
elif website == "youtube_channel":
    link="https://www.youtube.com/user/"+channel+"/search?query=" #sucht auf den youtube kanälen der ausgewählten youtuber
elif website == "medium":
    link= "https://medium.com/search?q=" # internes medium search
elif website == "google":
    link= "https://google.com/search?q=" # normal google search

'''
keyword: kannst du umändern 

link: Algorithmus 1 funktioniert nur mit links die als suche search_query an deren url anhängen - klappt bei google und medium
'''
```

```python
import requests


def create_html_query(keyword, link):
    '''
    return: string: query link
    '''

    query = link
    #query += search_query
    
    for subword in keyword.split(sep = " "):
        query+=subword
        query+="%20"

    query=query[:-3]
    return query


query = create_html_query(keyword, link)
print(query)

r = requests.get(query)
print(r)
```

```python
from bs4 import BeautifulSoup
import re
```

```python
html_page = r.content
soup = BeautifulSoup(html_page, 'html.parser')
```

```python
'''
Algorithmus 1 

zeigt alle texte die im direkten parent einen link haben

'''


texts = soup.find_all(text=True)

for t in texts:
    parent = t.find_parent(href=True)
    if parent is not None and len(t)>30:
        titel = str(t)
        titel = re.sub(r"\s+", ' ', titel)
        print("Video Titel: "+titel)
        print("Link:"+str(parent.get("href")))
        print("\n")
        print("\n")


    

```
